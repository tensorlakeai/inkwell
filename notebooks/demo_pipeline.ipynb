```python
from inkwell.pipeline import Pipeline
from inkwell.io import read_pdf_document, convert_page_to_image

from inkwell.utils.visualize import draw_box, visualize_image
```

```python
PDF_PATH = "https://pub-5dc4d0c0254749378ccbcfffa4bd2a1e.r2.dev/sample_ratings_report.pdf" 
document = read_pdf_document(PDF_PATH)
first_page = document.pages[0]
first_page_image = convert_page_to_image(first_page)
draw_box(first_page_image, layout=[])
```

```python
pipeline = Pipeline()
results = pipeline.process(PDF_PATH, pages_to_parse=[0])
```

```python
# Visualize the layouts for the first page
results_first_page = results.pages[0]
draw_box(first_page_image, layout=results_first_page.layout, show_element_type=True)
```

```python
images = results_first_page.image_fragments()
tables = results_first_page.table_fragments()
text_blocks = results_first_page.text_fragments()
```

```python
for fragment in images:
    visualize_image(fragment.content.image)
    print("-"*100)
```

```python
for fragment in tables:
    visualize_image(fragment.content.image)
    print(f"Data in table:\n{fragment.content.data}")
    print(f"Detection Score: {fragment.content.score}")
    print("-"*100)
```

```python
for text_block in text_blocks:
    visualize_image(text_block.content.image)
    print(f"Text in text block:\n{text_block.content.text}")
    print(f"Detection Score: {text_block.content.score}")
    print("-"*100)
```
